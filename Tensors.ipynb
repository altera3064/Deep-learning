텐서는 **스칼라, 벡터, 행렬의 상위 개념**이다.

- 스칼라 = 텐서 -> 0 랭크
- 벡터 = 텐서 -> 1랭크
- 행렬 = 텐서 -> 2랭크
- 이후는 3,4,5...n 랭크 텐서

한 마디로 **여러 차원을 가진 배열**을 말한다.

*랭크: 축의 개수를 말한다.

### Dimension 차원

여기서 "**차원**"은  

- 2차원 공간에서는 벡터가 2개의 성분을 가지니까 2차원 벡터,
- 3차원 공간에서는 벡터가 3개의 성분을 가지니까 3차원 벡터

텐서로 말하면 한 축의 길이를 차원이라고 말한다.

### Seed 시드

- Seed는 난수생성(random number generation)에 사용되는 출발점이다.

- 즉, seed 값을 설정하면 항상 같은 난수 결과를 얻을 수 있다.

```python
torch.manual_seed(1729)
r1 = torch.rand(2,2)
print("A random tensor: ")
print(r1)

r2 = torch.rand(2,2)
print("\nA different random tensor: ")
print(r2)

torch.manual_seed(1729)
r3 = torch.rand(2,2)
print("\nShold match r1 and r2")
print(r3)
```

```python
ones = torch.ones(2,3)
print(ones)

twos = torch.ones(2,3)
print(twos)

three = ones + twos
print(three)
print(three.shape)

# 같은 사이즈의 텐서끼리만 합, 곱이 가능하다.
r1 = torch.rand(2,3)
r2 = torch.rand(2,2)

r3 = r1 + r2
```

```python
r = torch.rand(2,2) - 0.5 * 2
print('a random matrix, r: ')
print(r)

print('\nAbsolute vale of r: ')
print(torch.abs(r))

print('\nInverse sine of r: ')
print(torch.asin(r))

print('\nDeterminant of r: ')
print(torch.det(r))

print('\nsingular value decomposition of r: ')
print(torch.svd(r))

print('\nAverage and standard Deviation of r: ')
print(torch.std_mean(r))

print('\nMaximum value of r: ')
print(torch.max(r))
```
